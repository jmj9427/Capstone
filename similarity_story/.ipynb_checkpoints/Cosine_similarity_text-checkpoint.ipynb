```python
# 2글자 이상 명사, 형용사, 동사 추출
def word_separating(story1) :
    result=[]
    for sto in story1 :      
        one_result = []
        words = twitter.pos(sto)
        for word in words:
            if word[1] in ["Noun", "Adjective","Verb"] and len(word[0]) >= 2 :
                one_result.append(word[0])
        one_result_a = " ".join(one_result)
        result.append(one_result_a)
    return(result)
```

```python
word_list = word_separating(story1) # word_list 생성
```

```python
# tf-idf
idf_maker = TfidfVectorizer(min_df=3)
```

```python
tf_idf = idf_maker.fit_transform(word_list).toarray()
```

```python
tf_idf_dtm = pd.DataFrame(tf_idf, columns= idf_maker.get_feature_names(), 
                      index= story['text'])
```

```python
# consine similarity
def cosin_similarity(doc1, doc2) :
    son = sum(doc1*doc2)
    mom = (sum((doc1)**2)*sum((doc2)**2))**0.5
    return(son/mom)
```

```python
def find_n_largest(doc_id, n):
    # 1번 문서와 가장 유사한 문서 4개 찾으려면 find_n_largest(1,4) 입력하면 됨!

    import numpy as np
    import pandas as pd

    # df=pd.read_excel("날짜_시간.xlsx")
    v = np.array(df[doc_id])

    v_max_index = list(np.argpartition(v, -n)[-n:])
    # 가장 큰 유사값들의 index를 list로 가져옴(해당 열에 중복 값 없다는 전제 하에)
    
    v_max = list(v[v_max_index])
    # 가장 큰 유사값들의 index를 사용하여 해당 값을 list로 가져옴
    
    new_df = pd.DataFrame({'doc':v_max_index,'doc_similarity':v_max})
    new_df.sort_values('doc_similarity',ascending=False,inplace=True)
    
    return (new_df)
    # index와 값을 이어서 데이터프레임으로 만든 것이 결과값으로 출력됨
    # 이것을 새로운 변수로 사용하려면,
    # 새로운 변수명 = find_n_largest(1,4) <- 이런식으로 사용하면됨
```

```python
songtitle=story['songtitle'] # 노래 추출
```

```python
similar_song=[]
```

```python
for i in range(0, 2128):
    a=find_n_largest(i,3)
    print(a)
    for j in range(0,3):
        similar_song.append(songtitle[a['doc'][j]])
```
