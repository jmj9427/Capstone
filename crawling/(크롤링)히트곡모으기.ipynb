## 주간차트에서 모든 히트곡들 노래 정보 모으기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import time
```

```python
song=pd.read_excel('1984-2018week_song.xlsx').drop_duplicates(subset='songid')
```

```python
song
```

```python
songebid=[]
songebnm=[]
songids=[]
singer=[]
title=[]
year=[]
genre=[]
like=[]
lyric=[]

url='https://www.melon.com/song/detail.htm?songId='
for sng in song['songid']:
    print(url+str(sng))
    r=requests.get(url+str(sng))
    sb=BeautifulSoup(r.text,'lxml')
    try:
        songebid.append(sb.select('#downloadfrm > div > div > div.entry > div.meta > dl > dd > a')[0].get('href').split('\'')[-2])
        songids.append(sng)
        songebnm.append(sb.select('#downloadfrm > div > div > div.entry > div.meta > dl > dd > a')[0].text)
        singer.append(sb.find('div',attrs={'class':'artist'}).text.strip())
        title.append(sb.select('div.song_name')[0].text.split('\t\t')[-5].strip())
        year.append(sb.select('#downloadfrm > div > div > div.entry > div.meta > dl > dd')[1])
        genre.append(sb.select('#downloadfrm > div > div > div.entry > div.meta > dl > dd')[2])
        lyric.append(sb.select('div.wrap_lyric')[0].text.strip())
    except AttributeError:
        next
    if len(songids) % 1000 == random.randrange(1,10):
        time.sleep(random.randrange(1,7))
```

```python
df=pd.DataFrame({'songebid':songebid,
                 'songebnm':songebnm,
                 'songid':songids,
                  'singer':singer,
                  'title':title,
                  'year':year,
                  'genre':genre,
                  'lyric':lyric
                 })
```

```python
df.to_excel('songinfo_melon_hit.xlsx',index=False)
```
