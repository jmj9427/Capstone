```python
url=''
text=[]
tdate=[]
title=[]
writer=[]
```

```python
for i in range(8122,8488):
    print(url+str(i))
    r=requests.get(url+str(i))
    sb=BeautifulSoup(r.text,'lxml')
    try:
        text.append(sb.find('div',attrs={'id':'content'}).text[46:].strip())
        tdate.append(sb.find('td',attrs={'headers':'con_date'}).text)
        title.append(sb.find('td',attrs={'headers':'con_title'}).text.strip())
        writer.append(sb.find('td',attrs={'headers':'con_writer'}).text.strip())
    except AttributeError:
        print('next')
        next
```

```python
pdf=pd.DataFrame({'radio':'nangmansidae',
              'date':tdate,
              'writer':writer,
              'title':title,
             'text':text})
pdf['len']=pdf['text'].apply(len)
pdf=pdf[pdf.len>350]
pdf.index=range(0,len(pdf))
pdf['textt']=pdf['title']+' '+pdf['text']
pdff=pdf[['date','radio','writer','textt']]
```

```python
pdff.to_excel('nangmansidae.xlsx',index=False)
```

```python
url='http://imbbs.imbc.com/view.mbc?list_id=5480003&pre_list_id=-1&next_list_id=5479987&page=1&bid=ablue142'
text=[]
writer=[]
tdate=[]
title=[]
for i in range(1,600):
    print(url)
    r=requests.get(url)
    r.encoding='euc-kr'
    sb=BeautifulSoup(r.text,'lxml')
    try:
        text.append(sb.find('span',attrs={'id':'rptView_ctl00_lblContent'}).text)
        title.append(sb.find('span',attrs={'id':'rptView_ctl00_lblTitle'}).text)
        writer.append(sb.select('#field_left')[0].text.strip().split('작성한 글 목록\n\n\n\n닫기\n\n\n\n\n\r\n\t\t\t\t                작성일 : ')[0])
        tdate.append(sb.select('#field_left')[0].text.strip().split('작성한 글 목록\n\n\n\n닫기\n\n\n\n\n\r\n\t\t\t\t                작성일 : ')[1])
    except AttributeError:
        next
    url='http://imbbs.imbc.com/'+sb.select('td > a')[-3].get('href')
```

```python
jdf=pd.DataFrame({'radio':'leedongjin',
              'date':tdate,
              'writer':writer,
              'title':title,
             'text':text})
jdf['len']=jdf['text'].apply(len)
jdf=jdf[jdf.len>350]
jdf.index=range(0,len(jdf))
jdf['textt']=jdf['title']+' '+jdf['text']
jdff=jdf[['date','radio','writer','textt']]
```

```python
jdff.to_excel('leedongjin.xlsx',index=False)
```

## MBC 메모장 형식

양요섭 : http://talkbbs.imbc.com/MemoList.asp?PG=2&SvcDiv=2759&IconNo=&FC=

```python
url1='http://talkbbs.imbc.com/MemoList.asp?PG='
url2='&SvcDiv=2759&IconNo=&FC='
text=[]
writer=[]
tdate=[]
title=[]
for i in range(1,209):
    print(url1+str(i)+url2)
    r=requests.get(url1+str(i)+url2)
    r.encoding='euc-kr'
    sb=BeautifulSoup(r.text,'lxml')
    for elm in sb.find_all('td',attrs={'width':'50%'}):
        writer.append(elm.text.strip())
    for elm in sb.find_all('td',attrs={'align':'right'})[4:-2]:
        tdate.append(elm.text.strip())
    for elm in sb.find_all('td',attrs={'width':'500'}):
        text.append(elm.text.strip())
```
