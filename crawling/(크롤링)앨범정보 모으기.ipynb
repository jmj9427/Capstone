```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import random
import time
```

```python
elbumlist=pd.read_excel('elbumlist_melon_hit.xlsx')
elbumlist=elbumlist.sort_index(by='songebid',ascending=False)
```

```python
songebids=[]
songebnm=[]
singer=[]
singerid=[]
year=[]
genre=[]
balmaeja=[]
ent=[]
text=[]
```

```python
url='https://www.melon.com/album/detail.htm?albumId='
for eb in elbumlist['songebid']:
    print(url+str(eb))
    r=requests.get(url+str(eb))
    sb=BeautifulSoup(r.text,'lxml')
    songebnm.append(sb.select('div.song_name')[0].text.replace('앨범명','').strip())
    try:
        singer.append(sb.select('div.artist > a')[0].text)
    except IndexError:
        singer.append('Varios Artists')
    try:
        singerid.append(sb.select('div.artist > a')[0].get('href').split('\'')[-2])
    except IndexError:
        singerid.append('none')
    year.append(sb.select('dl.list > dd')[0].text)
    genre.append(sb.select('dl.list > dd')[1].text)
    balmaeja.append(sb.select('dl.list > dd')[2].text)
    ent.append(sb.select('dl.list > dd')[3].text)
    text.append(sb.find('script',attrs={'type':'application/ld+json'}).text.split('description')[-1].split('potentialAction')[0].split('"')[-3])
    songebids.append(eb)
```

```python
elbuminfo=pd.DataFrame({'songebid':songebids,
                        'songebnm':songebnm,
                       'singer':singer,
                       'singerid':singerid,
                       'year':year,
                       'genre':genre,
                       'balmaeja':balmaeja,
                       'ent':ent,
                       'text':text})
```

```python
elbuminfo.to_excel('elbuminfo.xlsx',index=False)
```
